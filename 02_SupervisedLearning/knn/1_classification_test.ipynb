```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
# 定义数据
X = np.array([[2, 1], [3, 1], [1, 4], [2, 6]])
y = np.array([0, 0, 0, 1])  # 二分类

# 定义模型
# weights 参数指定距离的权重，'uniform' 表示所有邻居权重相等，'distance' 表示距离越近权重越大
# distance 参数指定距离度量方式，默认是欧氏距离（'minkowski' with p=2）
# 如果两个邻居类别相同，则预测该类别；如果不同，则根据权重决定类别
knn = KNeighborsClassifier(n_neighbors=2, weights='distance')
```

```python
# 训练模型
knn.fit(X, y)
```

```python
# 对新数据进行预测
print(knn.predict([[4, 9]]))
```

```python
import matplotlib.pyplot as plt
# 画图
X1 = X[y==0]
X2 = X[y==1]
```

```python
# 同一坐标轴上，每次会自动使用不同的颜色，这是由 Matplotlib 的 颜色循环（color cycle） 控制的。
plt.scatter(X1[:, 0], X1[:, 1])
plt.scatter(X2[:, 0], X2[:, 1])
plt.scatter(4, 9)

plt.show()
```
