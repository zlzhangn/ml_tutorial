# NumPy 高级操作和线性代数

本笔记本演示 NumPy 的高级操作，包括矩阵运算、广播和线性代数。

## 1. 广播（Broadcasting）

```python
import numpy as np

# 标量与数组
arr = np.array([1, 2, 3, 4, 5])
scalar = 10

print("数组:", arr)
print("标量:", scalar)
print()

print("arr + scalar:", arr + scalar)
print("arr * scalar:", arr * scalar)
print()
```

```python
# 不同形状数组的广播
arr_1d = np.array([[1, 2, 3], [4, 5, 6]])  # shape: (2, 3)
arr_2d = np.array([10, 20, 30])             # shape: (3,)

print("数组 (2x3):")
print(arr_1d)
print()

print("数组 (3,):")
print(arr_2d)
print()

print("广播后的加法:")
print(arr_1d + arr_2d)
```

## 2. 数组形状操作

```python
arr = np.arange(12)
print("原始数组:", arr)
print("形状:", arr.shape)
print()

# reshape - 改变形状，不改变数据
reshaped = arr.reshape(3, 4)
print("reshape(3, 4):")
print(reshaped)
print()

# flatten - 展平为1维
flattened = reshaped.flatten()
print("flatten():")
print(flattened)
print()

# ravel - 展平为1维（与原数组共享数据）
raveled = reshaped.ravel()
print("ravel():")
print(raveled)
```

```python
# transpose - 转置
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("原始数组 (2x3):")
print(arr)
print()

transposed = arr.T
print("转置后 (3x2):")
print(transposed)
```

## 3. 数组合并和分割

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("数组1:", arr1)
print("数组2:", arr2)
print()

# concatenate
concatenated = np.concatenate([arr1, arr2])
print("concatenate:", concatenated)
print()

# stack
stacked = np.stack([arr1, arr2])
print("stack:")
print(stacked)
```

```python
# 二维数组的合并
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

print("数组1:")
print(arr1)
print()

print("数组2:")
print(arr2)
print()

# vstack - 竖向堆积
vstack_result = np.vstack([arr1, arr2])
print("vstack:")
print(vstack_result)
print()

# hstack - 横向堆积
hstack_result = np.hstack([arr1, arr2])
print("hstack:")
print(hstack_result)
```

```python
# split - 分割数组
arr = np.array([1, 2, 3, 4, 5, 6])
print("原始数组:", arr)
print()

split_arrays = np.split(arr, 3)
print("split(arr, 3):")
for i, sub_arr in enumerate(split_arrays):
    print(f"  部分{i}:", sub_arr)
```

## 4. 排序和搜索

```python
arr = np.array([3, 1, 4, 1, 5, 9, 2, 6])
print("原始数组:", arr)
print()

# sort
sorted_arr = np.sort(arr)
print("排序:", sorted_arr)
print()

# argsort - 返回排序索引
indices = np.argsort(arr)
print("argsort:", indices)
print()

# searchsorted - 查找插入位置
sorted_arr = np.array([1, 2, 3, 4, 5])
print("排序数组:", sorted_arr)
print("searchsorted(3.5):", np.searchsorted(sorted_arr, 3.5))
```

```python
# unique - 去重
arr = np.array([1, 2, 2, 3, 3, 3, 4, 5, 5])
print("原始数组:", arr)
print()

unique_arr = np.unique(arr)
print("去重后:", unique_arr)
print()

# 包含计数
unique_arr, counts = np.unique(arr, return_counts=True)
print("去重值:", unique_arr)
print("计数:", counts)
```

## 5. 矩阵运算

```python
# 矩阵乘法
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print("矩阵 A:")
print(a)
print()

print("矩阵 B:")
print(b)
print()

# @ 运算符
result = a @ b
print("A @ B:")
print(result)
print()

# dot 函数
result = np.dot(a, b)
print("np.dot(A, B):")
print(result)
```

## 6. 线性代数运算

```python
# 行列式
a = np.array([[1, 2], [3, 4]])
print("矩阵 A:")
print(a)
print()

det = np.linalg.det(a)
print(f"行列式: {det}")
print()
```

```python
# 矩阵逆
a = np.array([[1, 2], [3, 4]])
print("矩阵 A:")
print(a)
print()

inv_a = np.linalg.inv(a)
print("A 的逆:")
print(inv_a)
print()

# 验证：A * A^-1 = I
print("A @ A^-1:")
print(a @ inv_a)
```

```python
# 求解线性方程组：Ax = b
a = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])

print("矩阵 A:")
print(a)
print()

print("向量 b:", b)
print()

x = np.linalg.solve(a, b)
print("解 x:", x)
print()

# 验证
print("验证 A @ x = b:")
print(a @ x)
```

```python
# 特征值和特征向量
a = np.array([[1, 2], [2, 1]])

print("矩阵 A:")
print(a)
print()

eigenvalues, eigenvectors = np.linalg.eig(a)

print("特征值:", eigenvalues)
print()

print("特征向量:")
print(eigenvectors)
```

```python
# SVD 分解
a = np.array([[1, 2, 3], [4, 5, 6]])

print("矩阵 A (2x3):")
print(a)
print()

U, S, Vt = np.linalg.svd(a)

print("U (2x2):")
print(U)
print()

print("S (奇异值):", S)
print()

print("V^T (3x3):")
print(Vt)
```

## 7. 范数和距离

```python
# 向量范数
v = np.array([3, 4])

print("向量:", v)
print()

# L2 范数（欧几里得距离）
norm_l2 = np.linalg.norm(v, ord=2)
print(f"L2 范数: {norm_l2}")
print()

# L1 范数（曼哈顿距离）
norm_l1 = np.linalg.norm(v, ord=1)
print(f"L1 范数: {norm_l1}")
print()

# 矩阵范数
a = np.array([[1, 2], [3, 4]])
matrix_norm = np.linalg.norm(a)
print(f"矩阵 Frobenius 范数: {matrix_norm}")
```
