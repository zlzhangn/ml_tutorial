# Matplotlib 高级图表

本笔记本演示 Matplotlib 的高级绘图功能，包括 3D 图、热力图、箱线图等。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
```

## 1. 箱线图（Box Plot）

```python
# 基础箱线图
fig, ax = plt.subplots(figsize=(10, 6))

# 创建示例数据
data = [np.random.normal(100, 15, 100) for _ in range(4)]

# 绘制箱线图
bp = ax.boxplot(data, labels=['A组', 'B组', 'C组', 'D组'],
                 patch_artist=True, widths=0.6)

# 美化箱线图
colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('数值', fontsize=12)
ax.set_title('箱线图示例', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
# 箱线图 + 散点图
fig, ax = plt.subplots(figsize=(10, 6))

# 创建示例数据
data = [np.random.normal(100, 15, 100) for _ in range(4)]

# 绘制箱线图
bp = ax.boxplot(data, labels=['A组', 'B组', 'C组', 'D组'],
                 patch_artist=True, widths=0.6)

# 美化
colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

# 添加散点
for i, d in enumerate(data):
    x = np.random.normal(i + 1, 0.04, size=len(d))
    ax.scatter(x, d, alpha=0.4, s=30, color='navy')

ax.set_ylabel('数值', fontsize=12)
ax.set_title('箱线图 + 散点图', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
## 2. 小提琴图（Violin Plot）
```

```python
# 小提琴图
fig, ax = plt.subplots(figsize=(10, 6))

# 创建示例数据
data = [np.random.normal(100, 15, 100) for _ in range(4)]

# 绘制小提琴图
parts = ax.violinplot(data, positions=[1, 2, 3, 4], widths=0.7,
                       showmeans=True, showmedians=True)

# 美化
for pc in parts['bodies']:
    pc.set_facecolor('skyblue')
    pc.set_alpha(0.7)

ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(['A组', 'B组', 'C组', 'D组'])
ax.set_ylabel('数值', fontsize=12)
ax.set_title('小提琴图示例', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
## 3. 热力图（Heatmap）
```

```python
# 基础热力图
fig, ax = plt.subplots(figsize=(10, 8))

# 创建数据矩阵
data = np.random.rand(8, 6) * 100

# 绘制热力图
im = ax.imshow(data, cmap='YlOrRd', aspect='auto')

# 添加颜色条
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('数值', fontsize=11)

# 设置标签
ax.set_xticks(np.arange(6))
ax.set_yticks(np.arange(8))
ax.set_xticklabels(['特征' + str(i) for i in range(1, 7)])
ax.set_yticklabels(['样本' + str(i) for i in range(1, 9)])

# 旋转标签
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

ax.set_title('热力图示例', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 相关系数热力图
fig, ax = plt.subplots(figsize=(10, 8))

# 创建数据框
np.random.seed(42)
data = np.random.randn(100, 5)
df = pd.DataFrame(data, columns=['特征1', '特征2', '特征3', '特征4', '特征5'])

# 计算相关系数矩阵
corr_matrix = df.corr()

# 绘制热力图
im = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto')

# 添加颜色条
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('相关系数', fontsize=11)

# 设置标签
ax.set_xticks(np.arange(len(corr_matrix)))
ax.set_yticks(np.arange(len(corr_matrix)))
ax.set_xticklabels(corr_matrix.columns, rotation=45, ha='right')
ax.set_yticklabels(corr_matrix.columns)

# 添加数值标签
for i in range(len(corr_matrix)):
    for j in range(len(corr_matrix)):
        text = ax.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}',
                       ha='center', va='center', color='black', fontsize=9)

ax.set_title('相关系数热力图', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
## 4. 3D 图表
```

```python
# 3D 曲线
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 创建参数
t = np.linspace(0, 4*np.pi, 100)
x = np.cos(t)
y = np.sin(t)
z = t

# 绘制 3D 螺旋线
ax.plot(x, y, z, 'b-', linewidth=2, label='螺旋线')

ax.set_xlabel('X 轴', fontsize=11)
ax.set_ylabel('Y 轴', fontsize=11)
ax.set_zlabel('Z 轴', fontsize=11)
ax.set_title('3D 螺旋线', fontsize=14, fontweight='bold')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# 3D 曲面
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 创建网格
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 绘制 3D 曲面
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8,
                        edgecolor='none')

# 添加颜色条
fig.colorbar(surf, ax=ax, label='Z 值')

ax.set_xlabel('X 轴', fontsize=11)
ax.set_ylabel('Y 轴', fontsize=11)
ax.set_zlabel('Z 轴', fontsize=11)
ax.set_title('3D 曲面: $z = sin(\\sqrt{x^2 + y^2})$', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 3D 柱状图
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 创建数据
xpos = np.array([1, 2, 3, 4, 5])
ypos = np.array([1, 2, 3])
zpos = np.zeros((3, 5))

dx = 0.8
dy = 0.8
dz = np.array([[10, 20, 30, 40, 50],
                [15, 25, 35, 45, 55],
                [12, 22, 32, 42, 52]])

# 创建颜色
colors = ['red', 'blue', 'green']
colors_extended = []
for color in colors:
    colors_extended.extend([color] * 5)

# 绘制 3D 柱状图
xpos_flat = np.repeat(xpos, 3)
ypos_flat = np.tile(ypos, 5)
zpos_flat = np.zeros(15)
dz_flat = dz.T.flatten()

ax.bar3d(xpos_flat - 1, ypos_flat - 1, zpos_flat, dx, dy, dz_flat,
        color=colors_extended, alpha=0.8)

ax.set_xlabel('X', fontsize=11)
ax.set_ylabel('Y', fontsize=11)
ax.set_zlabel('Z', fontsize=11)
ax.set_title('3D 柱状图', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
## 5. 等高线图（Contour Plot）
```

```python
# 等高线图
fig, ax = plt.subplots(figsize=(10, 8))

# 创建数据
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 绘制等高线
contour = ax.contour(X, Y, Z, levels=15, colors='black', linewidths=0.5)
ax.clabel(contour, inline=True, fontsize=8)

# 绘制填充等高线
contourf = ax.contourf(X, Y, Z, levels=20, cmap='viridis', alpha=0.8)

# 添加颜色条
cbar = plt.colorbar(contourf, ax=ax)
cbar.set_label('Z 值', fontsize=11)

ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_title('等高线图', fontsize=14, fontweight='bold')
ax.set_aspect('equal')

plt.tight_layout()
plt.show()
```

```python
## 6. 极坐标图
```

```python
# 极坐标线图
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(projection='polar'))

theta = np.linspace(0, 2*np.pi, 100)
r = 1 + 0.5 * np.sin(5*theta)

ax.plot(theta, r, linewidth=2, color='blue')
ax.fill(theta, r, alpha=0.3, color='blue')

ax.set_title('极坐标图: $r = 1 + 0.5sin(5\\theta)$',
            fontsize=14, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 极坐标柱状图
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(projection='polar'))

n = 12
theta = np.linspace(0, 2*np.pi, n, endpoint=False)
r = np.random.rand(n) * 10 + 5
width = 2*np.pi / n

colors = plt.cm.viridis(np.linspace(0, 1, n))
ax.bar(theta, r, width=width, alpha=0.8, color=colors, edgecolor='white', linewidth=2)

ax.set_title('极坐标柱状图', fontsize=14, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
