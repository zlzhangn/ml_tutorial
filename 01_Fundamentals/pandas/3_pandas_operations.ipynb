# Pandas 数据清洗和转换

本笔记本演示 Pandas 中的数据清洗和转换操作。

```python
import pandas as pd
import numpy as np

# 创建示例 DataFrame
df = pd.DataFrame({
    '日期': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    '产品': ['A', 'B', 'A', 'C', 'B'],
    '销售': [100, 200, 150, np.nan, 180],
    '利润': [10, 20, 15, 12, 18]
})

print("原始 DataFrame:")
print(df)
```

## 1. 处理缺失值

```python
# 检查缺失值
print("检查缺失值 (isnull):")
print(df.isnull())
print()

print("每列缺失值总数:")
print(df.isnull().sum())
print()

print("缺失值总数:")
print(df.isnull().sum().sum())
```

```python
# 删除缺失值
print("删除缺失值 (dropna):")
print(df.dropna())
print()

# 删除特定列中有缺失值的行
print("删除 '销售' 列缺失值的行:")
print(df.dropna(subset=['销售']))
```

```python
# 填充缺失值
print("用 0 填充缺失值 (fillna):")
print(df.fillna(0))
print()

# 用前向填充
print("前向填充 (ffill):")
print(df.fillna(method='ffill'))
print()

# 用后向填充
print("后向填充 (bfill):")
print(df.fillna(method='bfill'))
```

```python
# 用平均值填充
print("用平均值填充:")
mean_value = df['销售'].mean()
print(f"销售列平均值: {mean_value}")
print(df.fillna({'销售': mean_value}))
```

## 2. 数据类型转换

```python
# 创建示例数据
df2 = pd.DataFrame({
    '数字字符': ['1', '2', '3', '4'],
    '日期': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04']
})

print("原始数据类型:")
print(df2.dtypes)
print()
```

```python
# 转换为数值类型
df2['数字字符'] = df2['数字字符'].astype('int')
print("转换数字字符为 int:")
print(df2.dtypes)
print(df2)
print()
```

```python
# 转换为日期类型
df2['日期'] = pd.to_datetime(df2['日期'])
print("转换日期为 datetime:")
print(df2.dtypes)
print(df2)
print()
```

```python
# pd.to_numeric - 转换为数值，处理错误
df3 = pd.DataFrame({
    '数据': ['1', '2', '无效', '4']
})

print("原始数据:")
print(df3)
print()

# coerce 参数将无效值转为 NaN
df3['数据'] = pd.to_numeric(df3['数据'], errors='coerce')
print("转换后（无效值变为 NaN）:")
print(df3)
```

## 3. 重复值处理

```python
# 创建包含重复值的数据
df4 = pd.DataFrame({
    '姓名': ['Alice', 'Bob', 'Alice', 'David', 'Bob'],
    '年龄': [25, 30, 25, 35, 30]
})

print("原始 DataFrame（包含重复）:")
print(df4)
print()
```

```python
# 检查重复
print("duplicate_rows 检查:")
print(df4.duplicated())
print()

# 基于特定列检查重复
print("基于 '姓名' 列检查重复:")
print(df4.duplicated(subset=['姓名']))
print()

# drop_duplicates - 删除重复
print("删除重复行:")
print(df4.drop_duplicates())
print()

# 基于特定列删除重复
print("基于 '姓名' 删除重复（保留第一个）:")
print(df4.drop_duplicates(subset=['姓名'], keep='first'))
```

## 4. 行列操作

```python
# 重新设置索引
df5 = pd.DataFrame({
    '产品': ['A', 'B', 'C'],
    '销售': [100, 200, 150]
}, index=['产品1', '产品2', '产品3'])

print("原始 DataFrame:")
print(df5)
print()

# reset_index - 重置为整数索引
print("reset_index:")
print(df5.reset_index())
print()

# set_index - 设置新索引
print("set_index('产品'):")
print(df5.set_index('产品'))
```

```python
# 增加列和行
df6 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

print("原始 DataFrame:")
print(df6)
print()

# 添加列
df6['C'] = df6['A'] + df6['B']
print("添加列 C = A + B:")
print(df6)
print()

# 添加行
new_row = pd.DataFrame({'A': [4], 'B': [7], 'C': [11]})
df6 = pd.concat([df6, new_row], ignore_index=True)
print("添加新行后:")
print(df6)
```

## 5. 排序和排名

```python
# 创建示例数据
df7 = pd.DataFrame({
    '员工': ['Alice', 'Bob', 'Charlie', 'David'],
    '工作年数': [3, 7, 2, 5],
    '薪资': [50000, 70000, 45000, 65000]
})

print("原始 DataFrame:")
print(df7)
print()
```

```python
# sort_values - 按值排序
print("按薪资升序排序:")
print(df7.sort_values('薪资'))
print()

print("按薪资降序排序:")
print(df7.sort_values('薪资', ascending=False))
print()

# sort_index - 按索引排序
print("按索引降序排序:")
print(df7.sort_index(ascending=False))
```

```python
# rank - 排名
print("薪资排名:")
df7['薪资排名'] = df7['薪资'].rank(ascending=False)
print(df7)
print()

# 使用 'dense' 方法
df7['工作年数排名'] = df7['工作年数'].rank(method='dense', ascending=False)
print("工作年数排名 (dense):")
print(df7)
```

## 6. Apply 和 Map 函数

```python
# 创建示例数据
df8 = pd.DataFrame({
    '姓名': ['Alice', 'Bob', 'Charlie'],
    '分数': [85, 92, 78]
})

print("原始 DataFrame:")
print(df8)
print()
```

```python
# apply 应用自定义函数
def classify_score(score):
    """根据分数分类"""
    if score >= 90:
        return '优秀'
    elif score >= 80:
        return '良好'
    elif score >= 70:
        return '及格'
    else:
        return '不及格'

df8['等级'] = df8['分数'].apply(classify_score)
print("应用分类函数后:")
print(df8)
print()
```

```python
# apply 应用到整行
def get_info(row):
    """合并信息"""
    return f"{row['姓名']} ({row['等级']})"

df8['信息'] = df8.apply(get_info, axis=1)
print("应用行函数后:")
print(df8)
print()
```

```python
# map 映射值
grade_map = {90: 'A', 92: 'A', 85: 'B', 78: 'C'}
df8['字母成绩'] = df8['分数'].map(grade_map)
print("使用 map 映射:")
print(df8)
```
