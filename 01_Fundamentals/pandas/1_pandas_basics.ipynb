# Pandas 基础 - Series 和 DataFrame

本笔记本演示 Pandas 的基础概念，包括 Series、DataFrame 的创建和基本操作。

## 1. 导入 Pandas

```python
import pandas as pd
import numpy as np

# 检查版本
print(f"Pandas 版本: {pd.__version__}")
print(f"NumPy 版本: {np.__version__}")
```

## 2. Series 基础

Series 是一个一维数组，具有索引（标签）和数据。

```python
# 从列表创建 Series
s = pd.Series([10, 20, 30, 40, 50])
print("从列表创建的 Series:")
print(s)
print()

# Series 的属性
print(f"索引: {s.index.tolist()}")
print(f"值: {s.values}")
print(f"数据类型: {s.dtype}")
print(f"长度: {len(s)}")
```

```python
# 使用自定义索引创建 Series
s = pd.Series([10, 20, 30, 40, 50], 
               index=['a', 'b', 'c', 'd', 'e'])
print("使用自定义索引的 Series:")
print(s)
print()

# 按标签访问
print(f"s['a']: {s['a']}")
print(f"s[['a', 'c', 'e']]: ")
print(s[['a', 'c', 'e']])
```

```python
# 从字典创建 Series
data = {'apple': 100, 'banana': 150, 'orange': 120, 'grape': 200}
s = pd.Series(data)
print("从字典创建的 Series:")
print(s)
print()

# Series 的基本统计
print(f"平均值: {s.mean()}")
print(f"最大值: {s.max()}")
print(f"最小值: {s.min()}")
print(f"总和: {s.sum()}")
```

## 3. DataFrame 基础

DataFrame 是一个二维表格数据结构，类似于电子表格。

```python
# 从字典创建 DataFrame
data = {
    '名字': ['张三', '李四', '王五', '赵六'],
    '年龄': [25, 30, 35, 28],
    '城市': ['北京', '上海', '深圳', '杭州'],
    '工资': [15000, 18000, 20000, 17000]
}

df = pd.DataFrame(data)
print("从字典创建的 DataFrame:")
print(df)
print()

# DataFrame 的属性
print(f"形状: {df.shape}")
print(f"列名: {df.columns.tolist()}")
print(f"索引: {df.index.tolist()}")
print(f"数据类型:\n{df.dtypes}")
```

```python
# 访问列
print("访问单列（Series）:")
print(df['名字'])
print()

# 访问多列
print("访问多列（DataFrame）:")
print(df[['名字', '年龄']])
```

```python
# 访问行
print("访问第一行 (iloc[0]):")
print(df.iloc[0])
print()

print("访问多行 (iloc[0:2]):")
print(df.iloc[0:2])
```

```python
# loc 和 iloc 的区别
print("iloc - 基于位置的索引:")
print(df.iloc[0, 1])  # 第一行，第二列
print()

print("loc - 基于标签的索引:")
print(df.loc[0, '年龄'])  # 第0行，'年龄'列
print()
```

## 4. 从 NumPy 数组创建 DataFrame

```python
# 从 NumPy 数组创建
data = np.random.randn(4, 3)
df = pd.DataFrame(
    data,
    columns=['A', 'B', 'C'],
    index=['row1', 'row2', 'row3', 'row4']
)

print("从 NumPy 数组创建的 DataFrame:")
print(df)
print()

# 显示前几行
print("head() - 显示前5行:")
print(df.head())
print()

# 显示后几行
print("tail(2) - 显示后2行:")
print(df.tail(2))
```

## 5. DataFrame 的基本统计信息

```python
# describe() - 统计摘要
print("describe() - 统计摘要:")
print(df.describe())
print()
```

```python
# 列的统计函数
print("列 A 的统计函数:")
print(f"平均值: {df['A'].mean()}")
print(f"标准差: {df['A'].std()}")
print(f"最小值: {df['A'].min()}")
print(f"最大值: {df['A'].max()}")
print(f"求和: {df['A'].sum()}")
```

## 6. 添加和删除列

```python
# 添加列
df['D'] = [1, 2, 3, 4]
print("添加列 D 后:")
print(df)
print()
```

```python
# 基于现有列计算新列
df['E'] = df['A'] + df['B']
print("添加列 E (A + B) 后:")
print(df)
print()
```

```python
# 删除列
df_copy = df.copy()
df_drop = df_copy.drop('D', axis=1)  # 删除列 D
print("删除列 D 后:")
print(df_drop)
print()

# 删除行
df_drop_row = df_copy.drop(0, axis=0)  # 删除第一行
print("删除第一行后:")
print(df_drop_row)
```

## 7. 数据类型转换

```python
# 创建示例数据
df = pd.DataFrame({
    '整数': ['1', '2', '3', '4'],
    '浮点数': ['1.5', '2.5', '3.5', '4.5'],
    '类别': ['A', 'B', 'A', 'B']
})

print("原始 DataFrame:")
print(df)
print(df.dtypes)
print()
```

```python
# 类型转换
df['整数'] = df['整数'].astype('int')
df['浮点数'] = df['浮点数'].astype('float')
df['类别'] = df['类别'].astype('category')

print("转换后的 DataFrame:")
print(df)
print(df.dtypes)
```

## 8. 处理缺失值

```python
# 创建包含缺失值的数据
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, 7, 8],
    'C': ['x', 'y', 'z', 'w']
})

print("包含缺失值的 DataFrame:")
print(df)
print()
```

```python
# 检查缺失值
print("isnull() - 检查缺失值:")
print(df.isnull())
print()

print(f"缺失值总数: {df.isnull().sum().sum()}")
print()

# 每列缺失值个数
print("每列缺失值个数:")
print(df.isnull().sum())
```

```python
# 处理缺失值

# 1. 删除含有缺失值的行
print("dropna() - 删除含有缺失值的行:")
print(df.dropna())
print()

# 2. 填充缺失值
print("fillna(0) - 用 0 填充缺失值:")
print(df.fillna(0))
print()

# 3. 用平均值填充
print("fillna(mean()) - 用平均值填充:")
print(df.fillna(df.mean()))
```

## 9. 重命名列

```python
# 创建示例 DataFrame
df = pd.DataFrame({
    'col1': [1, 2, 3],
    'col2': [4, 5, 6],
    'col3': [7, 8, 9]
})

print("原始列名:")
print(df.columns.tolist())
print()

# 重命名特定列
df = df.rename(columns={'col1': 'A', 'col2': 'B', 'col3': 'C'})
print("重命名后:")
print(df.columns.tolist())
print(df)
```

## 10. 排序

```python
# 创建示例 DataFrame
df = pd.DataFrame({
    '名字': ['Alice', 'Bob', 'Charlie', 'David'],
    '年龄': [30, 25, 35, 28],
    '薪资': [50000, 45000, 60000, 48000]
})

print("原始 DataFrame:")
print(df)
print()
```

```python
# 按列排序
print("按年龄排序 (升序):")
print(df.sort_values('年龄'))
print()

print("按年龄排序 (降序):")
print(df.sort_values('年龄', ascending=False))
print()

# 按多列排序
print("按年龄升序、薪资降序排序:")
print(df.sort_values(['年龄', '薪资'], ascending=[True, False]))
```
