# Pandas GroupBy 和数据聚合

本笔记本演示 Pandas 中的 groupby 和数据聚合操作。

```python
import pandas as pd
import numpy as np

# 创建示例数据
data = {
    '部门': ['销售', '技术', '销售', '技术', '人力资源', '销售', '技术'],
    '姓名': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    '薪资': [50000, 65000, 55000, 70000, 48000, 52000, 68000],
    '年龄': [28, 35, 32, 40, 29, 31, 36],
    '绩效评分': [8, 9, 7, 9, 8, 8, 9]
}

df = pd.DataFrame(data)
print("原始数据:")
print(df)
```

## 1. 基本 GroupBy 操作

```python
# groupby 基本操作
print("按部门分组，计算薪资平均值:")
print(df.groupby('部门')['薪资'].mean())
print()

# groupby 返回一个对象
grouped = df.groupby('部门')
print(f"GroupBy 对象: {grouped}")
print(f"分组数: {grouped.ngroups}")
print(f"分组大小: {grouped.size()}")
```

## 2. 聚合函数

```python
# 单个聚合函数
print("各部门薪资总和:")
print(df.groupby('部门')['薪资'].sum())
print()

print("各部门薪资最大值:")
print(df.groupby('部门')['薪资'].max())
print()

print("各部门员工数:")
print(df.groupby('部门').size())
```

```python
# 多个聚合函数
print("各部门薪资统计:")
print(df.groupby('部门')['薪资'].agg(['mean', 'sum', 'min', 'max', 'count']))
print()

# 使用 agg 的字典形式
print("指定不同列的聚合函数:")
agg_dict = {
    '薪资': ['mean', 'sum'],
    '年龄': 'mean',
    '绩效评分': ['min', 'max']
}
print(df.groupby('部门').agg(agg_dict))
```

```python
# 多列分组
df2 = pd.DataFrame({
    '地区': ['北京', '北京', '上海', '上海', '深圳', '深圳'],
    '产品': ['A', 'B', 'A', 'B', 'A', 'B'],
    '销售': [100, 150, 120, 180, 110, 160]
})

print("原始数据:")
print(df2)
print()

print("按地区和产品分组，计算销售总和:")
print(df2.groupby(['地区', '产品'])['销售'].sum())
print()

# 转换为 DataFrame
print("以 DataFrame 形式显示:")
result = df2.groupby(['地区', '产品'])['销售'].sum().reset_index()
print(result)
```

## 3. Transform 和 Filter

```python
# transform - 返回原始形状的数据
print("原始数据:")
print(df[['部门', '薪资']].head())
print()

# 计算各部门平均薪资
df['部门平均薪资'] = df.groupby('部门')['薪资'].transform('mean')
print("添加部门平均薪资:")
print(df[['部门', '薪资', '部门平均薪资']].head())
print()

# 计算与部门平均的差异
df['薪资偏差'] = df['薪资'] - df['部门平均薪资']
print("薪资偏差:")
print(df[['部门', '薪资', '部门平均薪资', '薪资偏差']])
```

```python
# filter - 过滤分组
print("\n筛选人数 > 2 的部门:")
df3 = df.copy()
result = df3.groupby('部门').filter(lambda x: len(x) > 2)
print(result)
```

## 4. Pivot 和 Pivot Table

```python
# 创建示例数据
sales_data = pd.DataFrame({
    '日期': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02', 
             '2024-01-03', '2024-01-03'],
    '产品': ['A', 'B', 'A', 'B', 'A', 'B'],
    '销售': [100, 150, 120, 180, 110, 160]
})

print("原始数据:")
print(sales_data)
print()
```

```python
# pivot 表
print("Pivot 表（日期行，产品列）:")
pivot = sales_data.pivot(index='日期', columns='产品', values='销售')
print(pivot)
print()
```

```python
# pivot_table - 更灵活
print("Pivot Table（支持重复索引和聚合函数）:")
# 创建有重复的数据
data = {
    '地区': ['北京', '北京', '北京', '北京', '上海', '上海'],
    '产品': ['A', 'A', 'B', 'B', 'A', 'B'],
    '销售': [100, 110, 150, 160, 120, 180]
}
df = pd.DataFrame(data)
print("原始数据:")
print(df)
print()

print("Pivot Table（计算平均值）:")
pivot_table = df.pivot_table(
    values='销售',
    index='地区',
    columns='产品',
    aggfunc='mean'
)
print(pivot_table)
```

```python
## 5. 累积和滚动窗口
```

```python
# 创建时间序列数据
dates = pd.date_range('2024-01-01', periods=10)
df_ts = pd.DataFrame({
    '日期': dates,
    '销售': [100, 120, 110, 130, 140, 150, 145, 155, 160, 170]
})

print("原始数据:")
print(df_ts)
print()
```

```python
# 累积求和
df_ts['累积销售'] = df_ts['销售'].cumsum()
print("累积销售:")
print(df_ts[['日期', '销售', '累积销售']].head())
print()

# 累积最大值
df_ts['累积最大'] = df_ts['销售'].cummax()
print("累积最大:")
print(df_ts[['日期', '销售', '累积最大']].head())
```

```python
# 滚动平均
df_ts['滚动平均3'] = df_ts['销售'].rolling(window=3).mean()
print("3 日滚动平均:")
print(df_ts[['日期', '销售', '滚动平均3']])
print()

# 滚动求和
df_ts['滚动总和3'] = df_ts['销售'].rolling(window=3).sum()
print("3 日滚动总和:")
print(df_ts[['日期', '销售', '滚动总和3']])
```
