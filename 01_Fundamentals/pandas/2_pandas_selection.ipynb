# Pandas 数据选择和过滤

本笔记本演示 Pandas 中各种数据选择和过滤技术。

```python
import pandas as pd
import numpy as np

# 创建示例 DataFrame
data = {
    '员工ID': [101, 102, 103, 104, 105],
    '姓名': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    '部门': ['销售', '技术', '销售', '人力资源', '技术'],
    '年龄': [28, 35, 32, 29, 31],
    '薪资': [50000, 65000, 55000, 48000, 62000]
}

df = pd.DataFrame(data)
print("原始 DataFrame:")
print(df)
```

## 1. 访问列

```python
# 访问单列 - 返回 Series
print("访问 '姓名' 列:")
print(df['姓名'])
print()
print(f"类型: {type(df['姓名'])}")
```

```python
# 访问多列 - 返回 DataFrame
print("访问多列 ['姓名', '部门', '薪资']:")
print(df[['姓名', '部门', '薪资']])
print()
print(f"类型: {type(df[['姓名', '部门']])}")
```

```python
# 使用点号访问列（仅适用于有效的列名）
print("使用 df.姓名 访问列:")
print(df.姓名)
print()

# 获取所有列名
print("所有列名:")
print(df.columns.tolist())
```

## 2. 基于位置的访问 (iloc)

```python
# 访问单个元素
print("访问第 2 行，第 2 列的元素:")
print(df.iloc[1, 1])
print()

# 访问整行
print("访问第 2 行 (iloc[1]):")
print(df.iloc[1])
print()

# 访问整列
print("访问第 3 列 (iloc[:, 2]):")
print(df.iloc[:, 2])
```

```python
# 访问多行多列
print("访问前 3 行，前 3 列:")
print(df.iloc[0:3, 0:3])
print()

# 行切片
print("访问第 2 到 4 行:")
print(df.iloc[1:4])
print()

# 列切片
print("访问第 2 到 4 列:")
print(df.iloc[:, 1:4])
```

## 3. 基于标签的访问 (loc)

```python
# 访问单个元素
print("访问第 1 行，'姓名' 列的元素:")
print(df.loc[1, '姓名'])
print()

# 访问整行
print("访问索引为 2 的行:")
print(df.loc[2])
print()

# 访问特定列
print("访问 '薪资' 列:")
print(df.loc[:, '薪资'])
```

```python
# 访问多行
print("访问索引 1、2、4 的行:")
print(df.loc[[1, 2, 4]])
print()

# 访问行和列范围（包括两端）
print("访问索引 1 到 3 的行，'姓名' 到 '年龄' 列:")
print(df.loc[1:3, '姓名':'年龄'])
```

## 4. 布尔索引

```python
# 简单条件
print("薪资 > 50000 的员工:")
print(df[df['薪资'] > 50000])
print()

# 字符串条件
print("部门为 '技术' 的员工:")
print(df[df['部门'] == '技术'])
```

```python
# 多条件过滤（使用 & 和 |）
print("薪资 > 50000 且年龄 < 32 的员工:")
condition = (df['薪资'] > 50000) & (df['年龄'] < 32)
print(df[condition])
print()

print("部门为 '销售' 或 '技术' 的员工:")
condition = (df['部门'] == '销售') | (df['部门'] == '技术')
print(df[condition])
```

```python
# isin() 方法
print("部门在 ['销售', '技术'] 中的员工:")
print(df[df['部门'].isin(['销售', '技术'])])
print()

# 反向条件
print("薪资不超过 50000 的员工:")
print(df[~(df['薪资'] > 50000)])
```

## 5. 字符串方法

```python
# 创建包含字符串的示例数据
df2 = pd.DataFrame({
    '邮箱': ['alice@example.com', 'bob@company.com', 'charlie@example.com'],
    '姓名': ['Alice Smith', 'Bob Johnson', 'Charlie Brown']
})

print("原始 DataFrame:")
print(df2)
print()
```

```python
# 字符串方法
print("邮箱地址包含 'example' 的记录:")
print(df2[df2['邮箱'].str.contains('example')])
print()

print("姓名长度:")
print(df2['姓名'].str.len())
print()

print("姓名转大写:")
print(df2['姓名'].str.upper())
print()

print("姓名转小写:")
print(df2['姓名'].str.lower())
```

## 6. query() 方法

```python
# query() 使用字符串表达式进行过滤
print("使用 query 获取薪资 > 50000 的员工:")
print(df.query('薪资 > 50000'))
print()

print("使用 query 获取年龄在 28 到 32 之间的员工:")
print(df.query('28 <= 年龄 <= 32'))
print()

# 使用变量
min_salary = 55000
print(f"使用 query 获取薪资 >= {min_salary} 的员工:")
print(df.query('薪资 >= @min_salary'))
```

## 7. between() 方法

```python
# between() 检查值是否在范围内
print("年龄在 28 到 32 之间的员工:")
print(df[df['年龄'].between(28, 32)])
print()

print("薪资在 50000 到 60000 之间的员工:")
print(df[df['薪资'].between(50000, 60000)])
```

## 8. 条件替换

```python
# 创建副本用于演示
df_copy = df.copy()

# 条件替换值
df_copy.loc[df_copy['薪资'] > 60000, '薪资'] = 60000
print("薪资 > 60000 的改为 60000:")
print(df_copy)
print()

# 使用 where 方法
df_copy2 = df.copy()
df_copy2['薪资'] = df_copy2['薪资'].where(df_copy2['薪资'] <= 60000, 60000)
print("使用 where() 替换薪资 > 60000 为 60000:")
print(df_copy2)
```
