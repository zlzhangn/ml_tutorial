# Pandas 数据合并和连接

本笔记本演示 Pandas 中的数据合并、连接和拼接操作。

```python
import pandas as pd
import numpy as np
```

## 1. Concat - 拼接

```python
# 创建示例数据
df1 = pd.DataFrame({
    '姓名': ['Alice', 'Bob'],
    '年龄': [25, 30]
})

df2 = pd.DataFrame({
    '姓名': ['Charlie', 'David'],
    '年龄': [35, 28]
})

print("DataFrame 1:")
print(df1)
print()

print("DataFrame 2:")
print(df2)
print()
```

```python
# 纵向拼接（行）
print("纵向拼接 (axis=0):")
result = pd.concat([df1, df2], ignore_index=True)
print(result)
print()
```

```python
# 横向拼接（列）
df3 = pd.DataFrame({
    '城市': ['北京', '上海']
})

print("DataFrame 3:")
print(df3)
print()

print("横向拼接 (axis=1):")
result = pd.concat([df1, df3], axis=1)
print(result)
```
